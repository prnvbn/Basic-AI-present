### Linear Regression

1. Uses supervised learning i.e requires a labelled dataset
2. Used to fit the data to a linear equation

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1) 
```

```python
def plot_regression_line(x, y, b): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g") 
  
    # putting labels 
    plt.xlabel('x') 
    plt.ylabel('y') 
  
    # function to show plot 
    plt.show() 
```

```python
# generating dataset along the line 2x + 1 
x = np.linspace(-3, 3, 50)
y = 2*x +1 + np.random.uniform(-0.75, 0.5, 50)

plt.scatter(x, y)
plt.show()
```

```python
# estimating coefficients 
b = estimate_coef(x, y) 
print(f"Estimated line is {b[1]} *x + {b[0]}")

# plotting regression line 
plot_regression_line(x, y, b)
```
