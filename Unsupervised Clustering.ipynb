### K Means Clustering

1. Uses unsupervised learning i.e does not require labelled data.
2. Used to divide data into groups, however it does not label the groups.


```python
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Random dataset
data  = {
    'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
    'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]
}
plt.scatter(data['x'], data['y'])
plt.show()
```

```python
# Loading the dataset int a dataframe
df = pd.DataFrame(data=data, columns=['x', 'y'])
```

```python
# Training the model for 2 clusters
kmeans_model2  = KMeans(n_clusters=2)
model_info2    = kmeans_model2.fit(df)
centroids2     = model_info2.cluster_centers_
```

```python
# Plotting the clusters
plt.scatter(df['x'], df['y'], c=model_info2.labels_)
plt.scatter(centroids2[:, 0], centroids2[:, 1], c='red')
plt.show()
print(f"Centroid for the yellow group (cluster 1) is {centroids2[0]}")
print(f"Centroid for the purple group (cluster 2) is {centroids2[1]}")
```

```python
# Training the model for 3 clusters
kmeans_model3  = KMeans(n_clusters=3)
model_info3    = kmeans_model3.fit(df)
centroids3     = model_info3.cluster_centers_
```

```python
# Plotting the clusters
plt.scatter(df['x'], df['y'], c=model_info3.labels_)
plt.scatter(centroids3[:, 0], centroids3[:, 1], c='red')
plt.show()
print(f"Centroid for the yellow group (cluster 1) is {centroids3[0]}")
print(f"Centroid for the purple group (cluster 2) is {centroids3[1]}")
print(f"Centroid for the green  group (cluster 3) is {centroids3[2]}")
```
